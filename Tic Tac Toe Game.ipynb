## Implementation of Tic-Tac-Toe game

#### Rules of the Game

##### The game is to be played between two people (in this program between HUMAN and COMPUTER).
##### One of the player chooses ‘O’ and the other ‘X’ to mark their respective cells.
##### The game starts with one of the players and the game ends when one of the players has one whole row/ column/ diagonal filled with his/her respective character (‘O’ or ‘X’).
##### If no one wins, then the game is said to be draw.


```python

choices = [str(x+1) for x in range(9)]

# same as above
#for x in range (0, 9) :
#    choices.append(str(x + 1))

playerOneTurn = True
winner = False
draw = False

def printBoard() :
    print( '\n-------------')
    for i in range(0,9,3):
        print( '| ' + choices[i] + ' | ' + choices[i+1] + ' | ' + choices[i+2] + ' | ')
        print( '-------------')

empty_space = 9
while not winner and not draw:
    printBoard()

    if playerOneTurn :
        print( "Player 1:")
    else :
        print( "Player 2:")

    try:
        choice = int(input(">>Make a move by selecting one of the numbers shown in the board "))
    except:
        print("please enter a valid field")
        continue
    try:
        if choices[choice - 1] == 'X' or choices [choice-1] == 'O':
            print("illegal move, plase try again")
            continue
    except IndexError:
        print("\nEntered move exceeds the board size, please enter a valid move with in the board size")
        continue

    if playerOneTurn :
        choices[choice - 1] = 'X'
        empty_space-=1
    else :
        choices[choice - 1] = 'O'
        empty_space-=1

    playerOneTurn = not playerOneTurn

    if((choices[0] == choices[4] and choices[0] == choices[8]) or 
       (choices[2] == choices[4] and choices[4] == choices[6])) :
        winner = True
        printBoard()

    if not winner:
        for x in range (0, 3) :
            y = x * 3
            if (choices[y] == choices[(y + 1)] and choices[y] == choices[(y + 2)]) :
                winner = True
                printBoard()
                break
            if (choices[x] == choices[(x + 3)] and choices[x] == choices[(x + 6)]) :
                winner = True
                printBoard()
                break
        if not winner and empty_space == 0:
            draw = True
            printBoard()
          
         
if winner:
    print ("Player " + str(int(playerOneTurn + 1)) + " wins!\n")
elif draw:
    print("\nMatch is Drawn!")
```
